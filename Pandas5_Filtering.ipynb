# Filtering with help of pandas

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("SurveyData.xlsx")
df
```

### applying filter on Q2 column to find values = 'yes'

```python
df['Q2'] == 'yes'    # this will create a boolean array. True if value == 'yes' 
                     # else False
```

```python
myFilt = df["Q2"] == 'yes'    # assing boolean values to myFilt.

df[myFilt]                   # this will print the whole table with matched rows.
df.loc[myFilt]               # same as above line

df.loc[myFilt ,'Q2']         # only prints column Q2
                             # after filter we write column names which we want to show.

    # df[myFilt,'Q2']        error. user .loc[filter, [columns]]
```

# logical operators in filtering in pandas
### &   -  and operator 
### |     -  or operator

### second eg.
### searching all rows in Q1 having wildboar or blackbear in it

```python
# or opearator demo
myFilt2 = df['Q1'].str.contains('wildboar') | df['Q1'].str.contains('blackbear')

print("values containing wildboar or blackbear in column Q1\n")
df.loc[myFilt2, "Q1"]
```

```python
# and opeartor demo
myFilt2 = df['Q1'].str.contains('Wildboar') & df['Q1'].str.contains('blackbear')

print("values containing both wildboar and blackbear in column Q1\n")
df.loc[myFilt2, "Q1"]
```
