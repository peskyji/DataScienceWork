# adding a data frame to another using append

```python
myDict = {
            'Name': ['Ankit','Sahil'],
            'Item': ['Cutter','Rubber'],             # all values length must be equal
            'Region': ['Clement Town','G Road'],     # like here we have 2 values.
            'Units' : [10,11]
}

df2 = pd.DataFrame(myDict)
df2
```

```python
# adding df2 to df . (point to be noted names of columns in both are exactly same.)

df.append(df2)  # rows are added but the added rows have indexes from 0 which is not
                # correct.

# so we use ignore_index = True indside append.
```

```python
df = df.append(df2 , ignore_index = True)
df
```

### deleting index using conditions in dataFrame

```python
# we write filter & pass it to df and extract indexes from there and then pass indexes
# to the drop method.
```

#### example : delete all rows having item = Pencil

```python
filt1 = df['Item'] == 'Pencil'
indexes_to_be_deleted = df[filt1].index   # extracting indexes which satisfy the filter

df = df.drop(index = indexes_to_be_deleted)

df
```
