# using sqlalchemy to write df

```python
from sqlalchemy import create_engine
import pymysql
```

```python
engine = create_engine('mysql+pymysql://root:abcd@localhost/datascience')

# creating connection to data base.
# mysql = database your are using
# pymysql = DBAPI needed to connect to the database.
```

```python
# writing data to the database.

df4.to_sql(name = 'sample_bank' , con = engine , if_exists = 'replace' , index = False)
```
