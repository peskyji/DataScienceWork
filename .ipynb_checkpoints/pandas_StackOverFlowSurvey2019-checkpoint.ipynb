# Analysis on the above survey.

## Q.) There are several specified work challenges written in priority order (First,Second and third).<br>Find out which country faces which challenge the most.

```python
# lets have a look at the data
df['WorkChallenge'].head(20)
```

```python
'''
 step 1 : Find all the categories/Types of work challeges. By Splitting the 
          WorkChallenge into 3 different order.This will return a new data frame.
          
 step 2 : add df with country with this new data frame.
 
 step 3 : now group the dataframe based on country.
 
'''
```

```python
# creating a new data frame with only 2 columns.
df2 = df[['Country','WorkChallenge']]
df2
```

```python
# splitting the workChallenge column into 3 different columns.
tempDf = df2['WorkChallenge'].str.split(";" , expand = True)
tempDf
```

```python
# renaming new coumns.
tempDf = tempDf.rename(columns = {0:'1stP' , 1:'2ndP' , 2:'3rdP'})
tempDf
```

```python
# adding this above dataframe to df2.
df2 = df2.join(tempDf)
df2
```

```python
# removing WorkChallenge from df2 as it is of no use now.
df2.drop(columns = 'WorkChallenge' , inplace = True)
df2
```

```python
# grouping based on country.
workChallenge_grp = df2.groupby('Country')
```

```python
workChallenge_grp['1stP'].value_counts().loc['India']
```

```python
workChallenge_grp['2ndP'].value_counts().loc['India']
```

```python
workChallenge_grp['3rdP'].value_counts().loc['India']
```

## Q.) How many people in india knows python.

```python
# lets look at the data first
myDf = df[['Country','LanguageWorkedWith']]
myDf
```

```python
myDf['LanguageWorkedWith'] = myDf['LanguageWorkedWith'].str.lower()
myDf
```

```python
India_grp = myDf.groupby('Country').get_group('India')
India_grp
```

```python
x = India_grp['LanguageWorkedWith'].str.contains('python').value_counts(dropna = False)
x
```

```python
# changing index of above pandas series.
x = x.rename(index = {False : 'Don\'t know Python', 
                  True : 'Knows Python',
                  np.nan : 'No Response'
                 })

x
```

```python
x.plot(kind='pie', figsize = (6,6), title = "Python Survey")
```

## Q.) top 10 countries which knows python.

```python
myDf
```

```python
country_grp = myDf.groupby('Country')
```

```python
ans = country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('python').value_counts())
ans
```

```python
# if we want only those values which are true then we can use sum.
ans = country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('python').sum())
ans
```

```python
# top 10 countries which knows python well.

ans.nlargest(10)
```
